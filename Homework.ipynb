```python
import pandas as pd
```

```python
def count_top_names(year):
    df = pd.read_csv("yob"+str(year)+".txt", names = ["Name", "Gender", "Count"])
    return df.sort_values(by="Count", ascending=False)
count_top_names(1880)
```

```python
def count_top3(years):
    years_lst = list()
    for year in years:
        df = pd.read_csv("yob"+str(year)+".txt", names = ["Name", "Gender", "Count"])
        df['Year'] = year
        years_lst.append(df)
    df_res = pd.concat(years_lst)
    df_res = df_res.groupby(by="Name").sum()
    return df_res.sort_values(by="Count", ascending=False)    
count_top3([1900, 1950, 2000])
```

```python
def count_dynamics(years):
    years_lst = list()
    for year in years:
        df = pd.read_csv("yob"+str(year)+".txt", names = ["Name", "Gender", "Count"])
        df['Year'] = year
        years_lst.append(df)
    df_res = pd.concat(years_lst)
    return df_res.groupby(["Gender", "Year"]).sum()
count_dynamics([1900, 1950, 2000])
```
